```python
import matplotlib.pyplot as plt
plt.style.use('default')

import sys
sys.path.append("/cluster/home/yhgong/")

from python.modules.features.plot.plot2d import *
```

```python
simFile = SimFileModel(
    simPath="/lfs/data/yhgong/productionRun/mergerAGN/MHD/Cooling_Feedback_ContiRandom15",
    hdf5FilePrefix="perseus_merger_hdf5_plt_cnt",
    fileSterMyr=1
)
```

```python
simName: str = "MergerAGN MHD jet15"
```

### Los Velocity Dispersion

```python
fig, ax = Plot2d().losVelDispersion(
    axis="z",
    simFile=simFile,
    calculationInfo=LosDispersionCalculationInfoModel(
        timeMyr=2900,
        rBoxKpc=150,
    ),
    plotInfo=Plot2dInfoModel(
        title=f"Los {simName}"
    )
)
fig.show()
```

### Turbulence Velocity

```python
fig1, ax1 = Plot2d().velocityFiltering(
    field=VelocityFilteringField.turbVtotal,
    axis="z",
    simFile=simFile,
    calculationInfo=VelocityFilteringCalculationInfoModel(
        timeMyr=3000,
        rBoxKpc=150,
        bulkTurbFilteringMaxScale=40
    ),
    plotInfo=Plot2dInfoModel(
        title=f"Turb Vel {simName}",
        isLog=True,
        zlimMin=1e5,
        zlimMax=1e9
    )
)
```

### Turbulence Heating Vazza

```python
fig2, ax2 = Plot2d().turbulenceHeatingVazza(
    mode=TurbulenceHeatingVazzaMode.TurbVel,
    axis="z",
    simFile=simFile,
    calculationInfo=TurbulenceHeatingVazzaCalculationInfoModel(
        timeMyr=3000,
        rBoxKpc=150,
        bulkTurbFilteringMaxScale=40
    ),
    plotInfo=Plot2dInfoModel(
        title=f"Turb Heating {simName}",
        isLog=True,
        zlimMin=1e30,
        zlimMax=1e42
    )
)
```
